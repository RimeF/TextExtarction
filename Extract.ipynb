```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np
import keras_ocr
from img2table.document import Image
from img2table.ocr import EasyOCR
```

Extracting text from image from satellite

```python
#import and read the image
image_path = r"C:\Users\rimef\Bureau\eOnsight - Stage 2024 - Exercice technique - OCR\Genova.png"

img = cv2.imread(image_path)
```

```python
#import the pretrained model
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
#run the model on the image
results = pipeline.recognize([img])
```

```python
#display the results of the extracted text on the image
fig, ax = plt.subplots(figsize=(10, 10))
keras_ocr.tools.drawAnnotations(img, results[0], ax=ax)
ax.set_title('Keras OCR Result Example')
plt.show()
```

Extracting text from table
